```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('medals.csv')
```

```python
data = data.drop('participant_title', axis = 1)
data = data.drop('country_3_letter_code', axis = 1)
data.dropna(inplace = True)
```

```python
data.head()
```

```python
df = data[['discipline_title', 'athlete_full_name']].copy()
```

```python
Athletics = df.loc[df['discipline_title'] == 'Athletics']
Swimming = df.loc[df['discipline_title'] == 'Swimming']
Wrestling = df.loc[df['discipline_title'] == 'Wrestling']
Boxing = df.loc[df['discipline_title'] == 'Boxing']
Shooting = df.loc[df['discipline_title'] == 'Shooting']
Canoe_Sprint = df.loc[df['discipline_title'] == 'Canoe Sprint']
Gymnastics_Artistic = df.loc[df['discipline_title'] == 'Gymnastics Artistic']
```

```python
Athletics = Athletics.drop('discipline_title', axis =1)
Swimming = Swimming.drop('discipline_title', axis =1)
Wrestling = Wrestling.drop('discipline_title', axis =1)
Boxing = Boxing.drop('discipline_title', axis =1)
Shooting = Shooting.drop('discipline_title', axis =1)
Canoe_Sprint = Canoe_Sprint.drop('discipline_title', axis =1)
Gymnastics_Artistic = Gymnastics_Artistic.drop('discipline_title', axis =1)
```

```python
import json

atheletics = Athletics.to_json(orient='records', lines=False)
with open('atheletics.json', 'w') as json_file:
    json_file.write(atheletics)

swimming = Swimming.to_json(orient='records', lines=False)
with open('swimming.json', 'w') as json_file:
    json_file.write(swimming)

wrestling = Wrestling.to_json(orient='records', lines=False)
with open('wrestling.json', 'w') as json_file:
    json_file.write(wrestling)

boxing = Boxing.to_json(orient='records', lines=False)
with open('boxing.json', 'w') as json_file:
    json_file.write(boxing)

shooting = Shooting.to_json(orient='records', lines=False)
with open('shooting.json', 'w') as json_file:
    json_file.write(shooting)

canoe_Sprint = Canoe_Sprint.to_json(orient='records', lines=False)
with open('canoe_Sprint.json', 'w') as json_file:
    json_file.write(canoe_Sprint)

gymnastics_Artistic = Gymnastics_Artistic.to_json(orient='records', lines=False)
with open('gymnastics_Artistic.json', 'w') as json_file:
    json_file.write(gymnastics_Artistic)
```
